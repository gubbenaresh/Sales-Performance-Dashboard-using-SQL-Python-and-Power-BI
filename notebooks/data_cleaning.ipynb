### Clean Column Names

```python

df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df.columns = df.columns.str.replace("-","_")

df.columns
```

### Convert Date Columns

```python
df['order_date'] = pd.to_datetime(df['order_date'] , errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'] , errors='coerce')
```

### Convert Date Columns

```python
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_month_name'] = df['order_date'].dt.month_name()
df['order_quaterly'] = df['order_date'].dt.quarter
```

```python
df.isnull().sum()
```

```python
# checking null values 
df["postal_code"].isnull().sum()
```

```python
df["postal_code"] = df["postal_code"].fillna(0)
```

```python
# fixing null values droping null values 
df = df.dropna(subset=["order_date", "sales", "profit", "quantity"])
```

```python
# duplicates rows will removed 
df = df.drop_duplicates()
```

```python
df[['profit','sales']].head()
```

```python
df["profit_margin"] = round((df["profit"] / df["sales"]) * 100, 2)
```

```python
df["profit_margin"] = df["profit_margin"].replace([np.inf, -np.inf], 0)
df["profit_margin"] = df["profit_margin"].fillna(0)
```

```python
df["shipping_days"] = (df["ship_date"] - df["order_date"]).dt.days
```

```python
def sales_category(value):
    if value < 100:
        return "Low"
    elif value < 500:
        return "Medium"
    else:
        return "High"

df["sales_category"] = df["sales"].apply(sales_category)
```
